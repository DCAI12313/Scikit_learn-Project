```python
quality_cat = df.quality.unique()
quality_cat.sort()
qual_TSD = []
for i, quality in enumerate(quality_cat):
    qual_TSD.append([quality, df['total sulfur dioxide'].loc[df['quality'] == quality].mean()])
```

```python
#the above I calculate the average of total sulfur dioxide for each quality of wine,
#then below I obtain 7 lines, with for each of them the mean of total sulfur dioxide
#corresponding to each quality level
```

```python
df_qual_TSD = pd.DataFrame(qual_TSD, columns = ['Quality' , 'Mean TSD'])
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = "Quality", y = "Mean TSD", data = df_qual_TSD)
plt.show()
```

```python
#not too obvious but as the total surful dioxide decreases our quality slightly gets better
```

```python
#now transform data to categorical from numerical to better analyze it
#alcohol clumn transformation
"""
Low when less than 9.5.
Moderate between 9.5 and 11.
High between 11 and 12.5.
Very High when higher than 12.5.
"""
```

```python
def alcohol_cat(alcohol):
    if alcohol <= 9.5:
        return "Low"
    elif alcohol <= 11:
        return "Moderate"
    elif alcohol <= 12.5:
        return "High"
    else:
        return "Very High"

df['alcohol_category'] = df['alcohol'].apply(alcohol_cat)
df.sample(frac=1).head()
```

```python
#now I use this data and relate it to the quality of the wines: Crosstab Pandas Object 
```

```python
import numpy as np
plt.figure(figsize=(15,30))

cross = pd.crosstab(index=df['quality'],columns=df['alcohol_category'],normalize='index')
cross.plot.barh(stacked=True,rot=40,cmap='crest_r').legend(bbox_to_anchor=(1.0, 1.0))
plt.xlabel('% distribution per category')
plt.xticks(np.arange(0,1.1,0.1))
plt.title("Wine Quality each {}".format('alcohol_category'))
plt.show()
```

```python
#as the alcohol content increases, the quality of the wine increases too.
#but we are not certain 100%
```

```python
#Pearson and Heatmap formula
#to identify relationships between graphs and features
```

```python
plt.figure(figsize=(15,2))
sns.heatmap(df.select_dtypes(include='number').corr().iloc[[-1]],
            cmap='RdBu_r', annot=True, vmin=-1, vmax=1)
```

```python
#we can see that the closer it is to either 1 or -1 the more positive or negative the correlations are, the most correlated features:
#alcohol, density chlorides, and volatile acidity
```

```python
df_label = df['quality']
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_features, df_label, test_size=0.20)
```

```python
print((len(X_train),len(y_train)))
print((len(X_test),len(y_test)))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logisticRegression = LogisticRegression()
logisticRegression.fit(X_train, y_train)
```

```python
logisticRegression.score(X_test, y_test)
```

```python
logisticRegression.predict_proba(X_test.iloc[:1])
#this is for the probabilities of belonging to each class for the first row of our dataset
```

Support Vector Machines

```python
from sklearn import svm

SVM = svm.SVC()
SVM.fit(X_train, y_train)
SVM.score(X_test, y_test)
```

Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDClassifier

SGD = SGDClassifier()
SGD.fit(X_train, y_train)
SGD.score(X_test, y_test)
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(X_train, y_train)
GNB.score(X_test, y_test)
```

Decision Tree

```python
from sklearn import tree

decisionTree = tree.DecisionTreeClassifier()
decisionTree.fit(X_train, y_train)
decisionTree.score(X_test, y_test)
```
